# Created a similar excel file with the columns "OrderDate" and "salesMan"

```python
# Let's first read the Excel file using pandas to see what it contains
import pandas as pd

# Load a sheet into a DataFrame
df = pd.read_excel('sales_data.xlsx')

df.head(20)  # display the first few rows of the dataframe
```

### Now, Replacing all empty cells in “SalesMan” column with my Name.

```python
df['SalesMan'].fillna('Kanchan Sharma', inplace=True)
df
```

```python
# viewing the dates of OrderDate column
df['OrderDate']
```

### Replace all cells in column “OrderDate” with date as “DD/MM/YY” to “DD/MM/20YY”

```python
# (this cell should not be run twice, as it modifies the actual df)

# Convert 'OrderDate' column to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'], format='%d/%m/%y')

# Format dates as 'DD/MM/20YY'
df['OrderDate'] = df['OrderDate'].apply(lambda x: x.strftime('%d/%m/20%y'))


df['OrderDate']
```
